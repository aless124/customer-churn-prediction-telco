```python
import pandas as pd
 import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Affichage des graphiques dans le notebook
```

## Chargement des données

```python
# Remplacer le chemin par le chemin réel du fichier téléchargé depuis Kaggle
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

## Analyse des valeurs manquantes

```python
df.isnull().sum().sort_values(ascending=False)
```

## Visualisation de la variable cible (churn)

```python
sns.countplot(x='Churn', data=df)
plt.title('Répartition de la variable cible (Churn)')
plt.show()
```

## Visualisation de quelques variables explicatives

```python
# Exemple : distribution de l'âge
if 'SeniorCitizen' in df.columns:
    sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
    plt.title('Churn selon SeniorCitizen')
    plt.show()
```

```python
# Exemple : distribution du type de contrat
if 'Contract' in df.columns:
    sns.countplot(x='Contract', hue='Churn', data=df)
    plt.title('Churn selon le type de contrat')
    plt.show()
```
